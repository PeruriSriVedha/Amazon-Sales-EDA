```python
df = df.drop_duplicates()
```

```python
df = df.drop(['New', 'PendingS'], axis = 1)
```

```python
unique_currencies = df['currency'].unique()
unique_currencies
```

```python
df = df.drop(['currency'], axis = 1)
```

```python
missing_ship_info = df[df['ship-city'].isnull() | df['ship-state'].isnull() | df['ship-postal-code'].isnull() | df['ship-country'].isnull()]

# Check if all missing values are in the same rows
all_missing_same_rows = missing_ship_info['ship-city'].isnull().equals(missing_ship_info['ship-state'].isnull()) and \
                       missing_ship_info['ship-city'].isnull().equals(missing_ship_info['ship-postal-code'].isnull()) and \
                       missing_ship_info['ship-city'].isnull().equals(missing_ship_info['ship-country'].isnull())

print(f"\nAll missing ship information values are in the same rows: {all_missing_same_rows}")
```

```python

df = df.dropna(subset=['ship-city', 'ship-state', 'ship-postal-code', 'ship-country'])
df
```

```python
unique_fulfilled = df['fulfilled-by'].unique()
unique_fulfilled
```

```python
df['fulfilled-by'] = df['fulfilled-by'].fillna('Normal')
```

```python
# Group data by specified columns
grouped = df.groupby(['Fulfilment', 'Sales Channel', 'ship-service-level', 'Category', 'Size', 'ship-city', 'B2B', 'fulfilled-by'])

# Function to replace NaN values with the average of the group
def replace_nan_with_group_average(group):
    average = group['Amount'].mean()
    group['Amount'] = group['Amount'].fillna(average)
    return group

# Apply the function to each group
df = grouped.apply(replace_nan_with_group_average).reset_index(drop=True)

# Round 'Amount' column to 2 decimal places
df['Amount'] = df['Amount'].round(2)
```

```python
df.head()
```

```python
df = df.sort_values(by = 'index')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```
